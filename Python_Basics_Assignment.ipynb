**Q 2. What is an interpreter in Python?**

Python runs on an interpreter system, meaning that code can be executed as soon as it is written. This means that prototyping can be very quick.

- **A Python Interpreter** is the program that reads and executes Python code. It translates Python instructions into machine-readable form line by line, so the computer can understand and run them.

  - When we run a Python program, code passes through Python Interpreter, which is responsible for checking your code for errors converting it into an intermediate form called bytecode, Sending it to the Python Virtual Machine (PVM) for execution.

**Q 11.	What is the purpose of conditional statements in Python?**

**Q 12.	How does the elif statement work?**

  -	The elif statement is a conditional statement which stand for else if. It is used to check the multiple condition. If the first condition is false, it moves on to the next “elif” statement to check if that condition is true.
  -	If the condition is true, the if block code is executed and if the condition is false, the else block code is executed.
  -	If there are n numbers of elif and first elif condition get matched then code will not check next condition.


**Q 13.	What is the difference between for and while loops?**

  Both for loops and while loops are control flow structures in programming that allow you to repeatedly execute a block of code. However, they differ in their syntax and use cases.

  -	**For loop**:

      -	It iterate over a sequence of elements and executes the block of code for each item in the sequence.

  -	**While loop**:

      -	It repeatedly execute block of code until a certain condition is met.

```python
#Q 9.	Write a Python program to convert user input from string to integer, float, and Boolean types.

#As we know user input bydefault string in nature so we have to do typecasting of user input

# Take input from user
user_input = input("Enter something: ")

# Convert to integer
try:
    int_value = int(user_input)
    print("Integer value:", int_value)
except ValueError:
    print("Cannot convert to integer.")

# Convert to float
try:
    float_value = float(user_input)
    print("Float value:", float_value)
except ValueError:
    print("Cannot convert to float.")

# Convert to Boolean
bool_value = bool(user_input)
print("As Boolean:", bool_value)
```

```python
#Q 10.	Write code to demonstrate type casting with list elements.

# A list of strings that look like numbers
my_list = ["10", "20", "30"]

# Convert all elements to an integer
int_list = [int(x) for x in my_list]
print("List as integers:", int_list)

# Convert all elements to a float
float_list = [float(x) for x in my_list]
print("List as floats:", float_list)

# Convert all elements to a float
str_list = [str(x) for x in my_list]
print("List as strings:", str_list)

# Convert all elements to a boolean
bool_list = [bool(x) for x in my_list]
print("List as booleans:", bool_list)
```

```python
#Q 14.	Write a program to reverse a string using a while loop.

text = "ABCDEF"
reversed_text = ""
i = len(text) - 1
while i >= 0:
    reversed_text += text[i]
    i -= 1
print("Reversed string:", reversed_text)
```

```python
#Q 15.	Write a Python program to calculate the factorial of a number provided by the user using a while loop

num = int(input("Enter a number: "))

factorial = 1
i = 1
while i <= num:
    factorial = factorial * i
    i = i+1
print("The factorial of number is ", factorial)
```
